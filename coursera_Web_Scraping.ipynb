```python

!pip install requests beautifulsoup4

import requests
from bs4 import BeautifulSoup
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
data = requests.get(url).text
print(data)
```

```python

soup = BeautifulSoup(data, 'html.parser')
```

```python
soup.find('table') #affiche le premier tableau trouv
```

```python
tables = soup.find_all('table')
# Accder aux autres tableaux
for index, table in enumerate(tables):
    print(f"Tableau {index + 1}:")
    print(table.prettify())  #
```

```python
second_table = tables[1]  # Les indices commencent  0
print("Deuxime Tableau:")
print(second_table.prettify())
```

```python
soup.find("table").find_all('tr')
```

```python
import pandas as pd
TESLA = pd.DataFrame(columns=["Year","Revenues"])
```

```python

for row in soup.find("tbody").find_all('tr'): 
    col = row.find_all("td")
    year = col[0].text
    revenues = col[1].text
    
   
    TESLA = pd.concat([TESLA,pd.DataFrame({"Year":[year], "Revenues":[revenues]})], ignore_index=True)    
```

```python
TESLA.tail()
```

```python
import pandas as pd
import re

# Fonction pour enlever le symbole dollar et les virgules
def remove_dollar(value):
    # Enlever les symboles $ et les virgules, puis retirer les espaces
    value = value.replace("$", "").replace(",", "").strip()
    return value 

TESLA['Revenues'] = TESLA['Revenues'].apply(remove_dollar)

# Convertir en numrique en grant les erreurs
TESLA['Revenues'] = pd.to_numeric(TESLA['Revenues'], errors='coerce')

```

```python

TESLA = TESLA.sort_values(by='Year')
```

```python
TESLA.tail()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crer le graphique
plt.figure(figsize=(10, 6))
sns.barplot(data=TESLA, x='Year', y='Revenues', hue = 'Year', palette='viridis',legend=False)

# Ajouter des titres et des labels
plt.title('Revenus Tesla per year', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Revenus', fontsize=14)

# Afficher le graphique
plt.show()
```
